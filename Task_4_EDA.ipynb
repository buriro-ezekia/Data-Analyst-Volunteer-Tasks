```python
# Summary statistics for key performance metrics
performance_metrics = df[['impressions', 'taps', 'installs', 'spend', 'average_cpt',
                          'average_cpa', 'average_cpm',
                          'ttr', 'cr']]

# Calculate and round to 4 decimal places
summary_stats = performance_metrics.describe().round(4)
summary_stats
```

```python
# Distribution Analysis
# Visualise the distribution of key metrics like impressions, taps, and conversions


import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Plot distributions
fig, axes = plt.subplots(3, 2, figsize=(15, 12))

# Impressions
sns.histplot(df['impressions'], kde=True, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Distribution of Impressions')

# Taps
sns.histplot(df['taps'], kde=True, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Distribution of Taps')

# Installs
sns.histplot(df['installs'], kde=True, ax=axes[1, 0], color='red')
axes[1, 0].set_title('Distribution of Installs')

# Spend
sns.histplot(df['spend'], kde=True, ax=axes[1, 1], color='purple')
axes[1, 1].set_title('Distribution of Spend')

# Average Cost per Tap (CPT)
sns.histplot(df['average_cpt'], kde=True, ax=axes[2, 0], color='orange')
axes[2, 0].set_title('Distribution of Average Cost per Tap (CPT)')

# Conversion Rate (CR)
sns.histplot(df['cr'], kde=True, ax=axes[2, 1], color='brown')
axes[2, 1].set_title('Distribution of Conversion Rate (CR)')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

- Most key metrics (impressions, taps, installs, and spend) are heavily right-skewed, indicating that a small number of campaigns are driving most of the performance.
- Skewness in spend and performance suggests that optimising budget allocation and improving underperforming campaigns could lead to more balanced and consistent results.
- High variability in CPT and CR indicates that some campaigns are more efficient than others, highlighting the need for targeting and creative adjustments.

```python
# Correlation Analysis

# Compute and visualise correlations between different performance metrics to understand their relationships

# Correlation matrix
correlation_matrix = performance_metrics.corr()

# Plot correlation heatmap
plt.figure(figsize=(9, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Performance Metrics')
plt.show()
```

The matrix uses a colour gradient from blue to red, where blue and red represent negative and positive correlations, respectively. The intensity of the colour indicates the strength of the correlation.

- Impressions, taps, and installs show very high correlations with each other (0.99 to 1). And now, this suggests that as the number of impressions increases, taps and installs also increase proportionally.

- Spending has a strong positive correlation with impressions, taps, and installs (all 0.94), indicating that higher spending is likely associated with higher values in these metrics.

- There is a slight negative correlation between spend and average cost per tap (-0.0085), average cost per acquisition (-0.0085), and average cost per thousand impressions (-0.11). That could imply that higher spending does not necessarily increase these cost metrics proportionally, possibly due to economies of scale.

- Average cost per tap, acquisition, and thousand impressions are highly correlated (0.75 to 0.89). It indicates that these costs tend to increase or decrease together.

- Conversion rate shows moderate positive correlations with taps (0.24), installs (0.28), and spend (0.28). It also correlates with the tap-through rate (0.59), suggesting that higher engagement and spending could lead to better conversion rates.



```python
# Campaign Performance Visualization
# Create plots to visualise and compare the performance of different campaigns

import warnings
warnings.filterwarnings('ignore')

# Plot campaign performance metrics

# Set the plot style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Impressions by Campaign
sns.barplot(x='campaign_name', y='impressions', data=df, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Impressions by Campaign')
axes[0, 0].set_xticklabels(df['campaign_name'], rotation=90, fontsize=10)

# Taps by Campaign
sns.barplot(x='campaign_name', y='taps', data=df, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Taps by Campaign')
axes[0, 1].set_xticklabels(df['campaign_name'], rotation=90, fontsize=10)

# Installs by Campaign
sns.barplot(x='campaign_name', y='installs', data=df, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Installs by Campaign')
axes[1, 0].set_xticklabels(df['campaign_name'], rotation=90, fontsize=10)

# Spend by Campaign
sns.barplot(x='campaign_name', y='spend', data=df, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Spend by Campaign')
axes[1, 1].set_xticklabels(df['campaign_name'], rotation=90, fontsize=10)

# Adjust layout to prevent overlapping
plt.tight_layout(pad=2.0)
plt.show()
```

Compute Performance Metrics to identify the top-performing and underperforming campaigns based on metrics like taps, conversions, and cost-per-acquisition (CPA)

```python
# Compute metrics
df['tap_to_conversion_rate'] = df['installs'] / df['taps']
df['cpa'] = df['average_cpa']
df['cpm'] = df['average_cpm']
df['ttr'] = df['taps'] / df['impressions']
```

```python
# Sort and rank
top_taps = df.sort_values(by='taps', ascending=False).head(3)
bottom_taps = df.sort_values(by='taps').head(3)

top_conversions = df.sort_values(by='installs', ascending=False).head(3)
bottom_conversions = df.sort_values(by='installs').head(3)

top_cpa = df.sort_values(by='cpa').head(3)  # Lower CPA is better
bottom_cpa = df.sort_values(by='cpa', ascending=False).head(3)
```

```python
# Print results for reference
print("Top Performing Campaigns by Taps:")
print(top_taps[['campaign_name', 'taps']])

print("\nUnderperforming Campaigns by Taps:")
print(bottom_taps[['campaign_name', 'taps']])

print("\nTop Performing Campaigns by Conversions:")
print(top_conversions[['campaign_name', 'installs']])

print("\nUnderperforming Campaigns by Conversions:")
print(bottom_conversions[['campaign_name', 'installs']])

print("\nTop Performing Campaigns by CPA (Low CPA):")
print(top_cpa[['campaign_name', 'cpa']])

print("\nUnderperforming Campaigns by CPA (High CPA):")
print(bottom_cpa[['campaign_name', 'cpa']])
```

```python
# Plot results
fig, axes = plt.subplots(3, 2, figsize=(15, 18))

# Top Taps
sns.barplot(x='campaign_name', y='taps', data=top_taps, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Top 3 Campaigns by Taps')
axes[0, 0].set_xticklabels(axes[0, 0].get_xticklabels(), rotation=45)

# Bottom Taps
sns.barplot(x='campaign_name', y='taps', data=bottom_taps, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Bottom 3 Campaigns by Taps')
axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=45)

# Top Conversions
sns.barplot(x='campaign_name', y='installs', data=top_conversions, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Top 3 Campaigns by Conversions')
axes[1, 0].set_xticklabels(axes[1, 0].get_xticklabels(), rotation=45)

# Bottom Conversions
sns.barplot(x='campaign_name', y='installs', data=bottom_conversions, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Bottom 3 Campaigns by Conversions')
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45)

# Top CPA
sns.barplot(x='campaign_name', y='cpa', data=top_cpa, ax=axes[2, 0], palette='viridis')
axes[2, 0].set_title('Top 3 Campaigns by Cost-Per-Acquisition (Low CPA)')
axes[2, 0].set_xticklabels(axes[2, 0].get_xticklabels(), rotation=45)

# Bottom CPA
sns.barplot(x='campaign_name', y='cpa', data=bottom_cpa, ax=axes[2, 1], palette='viridis')
axes[2, 1].set_title('Bottom 3 Campaigns by Cost-Per-Acquisition (High CPA)')
axes[2, 1].set_xticklabels(axes[2, 1].get_xticklabels(), rotation=45)

plt.tight_layout()
plt.show()
```

# Analysing trends in user acquisition and in-app purchase data

```python
# User Acquisition Analysis

# Group by 'start_date' and calculate sum of 'installs', 'impressions', 'taps',
# and average of 'conversion_rate_(cr)'
user_acquisition_data = df.groupby('start_date').agg({
    'installs': 'sum',
    'impressions': 'sum',
    'taps': 'sum',
    'cr': 'mean'
}).reset_index()

# Plotting
plt.figure(figsize=(12, 6))

# Plot Installs
plt.plot(user_acquisition_data['start_date'], user_acquisition_data['installs'], label='Installs', marker='o')

# Plot Impressions
plt.plot(user_acquisition_data['start_date'], user_acquisition_data['impressions'], label='Impressions', marker='o')

# Plot Taps
plt.plot(user_acquisition_data['start_date'], user_acquisition_data['taps'], label='Taps', marker='o')

plt.title('User Acquisition Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Summary

- High impressions but low engagement and conversion suggest issues with creative content and user targeting.
- Focus on improving ad relevance, refining targeting, and enhancing post-click experience.

```python
# In-App Purchase Analysis

# Group by 'start_date' and calculate sum of 'spend', 'installs', 'new_downloads', and 'redownloads'
in_app_purchase_data = df.groupby('start_date').agg({
    'spend': 'sum',
    'installs': 'sum',
    'new_downloads': 'sum',
    're_downloads': 'sum'
}).reset_index()

# Calculate total revenue
total_revenue = in_app_purchase_data['spend'].sum()

# Calculate average revenue per user
average_revenue_per_user = total_revenue / in_app_purchase_data['installs'].sum()

# Plotting
plt.figure(figsize=(12, 6))

# Plot Spend
plt.plot(in_app_purchase_data['start_date'], in_app_purchase_data['spend'], label='Spend', marker='o')

# Plot Installs
plt.plot(in_app_purchase_data['start_date'], in_app_purchase_data['installs'], label='Installs', marker='o')

# Plot New Downloads
plt.plot(in_app_purchase_data['start_date'], in_app_purchase_data['new_downloads'], label='New Downloads', marker='o')

# Plot Redownloads
plt.plot(in_app_purchase_data['start_date'], in_app_purchase_data['re_downloads'], label='Redownloads', marker='o')

plt.title('In-App Purchase Trends Over Time')
plt.xlabel('Date')
plt.ylabel(' Total ()')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```

**Summary of Trends**  
- Around **2020-01-01**, New Downloads (3,100), Spend (1,5001,700), and Installs peaked.  
- After the peak, all metrics dropped significantly by **2020-04-15**.  
- Redownloads remained consistently low throughout the period.  
- A smaller rise in Spend and New Downloads occurred from **2020-10-15** to **2020-12**.  
- All metrics dropped to low levels in **2021** and **2022**, indicating reduced app engagement and purchasing activity.  

**Recommendations**  
1. **Campaign Analysis:** Investigate the factors behind the successful 2020-01-01 campaign to replicate its success.  
2. **Retention Strategy:** Develop user retention strategies to maintain engagement post-campaign (e.g., push notifications, loyalty programmes).  
3. **Targeting Improvement:** Improve ad targeting and creative content to drive higher engagement during future campaigns.  
4. **Reactivation Campaign:** Launch campaigns focused on re-engaging past users and encouraging redownloads.  
5. **Data-Driven Adjustments:** Continuously monitor performance and adjust based on user response to maximise ROI.

```python
# Correlation Analysis
# Use Pearson correlation coefficient as it measures the linear relationship between two variables

# Select relevant columns for correlation analysis
correlation_data = df[['installs', 'impressions', 'taps', 'cr',
                       'spend', 'new_downloads', 're_downloads']]

# Calculate correlation matrix
correlation_matrix = correlation_data.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Matrix of User Acquisition and In-App Purchase Metrics')
plt.show()
```

**Key observations from the correlation matrix**

The colour scale on the right side of the matrix ranges from blue to red, where blue represents lower correlation values, and red represents higher correlation values.

- Metrics such as installs, impressions, taps, new downloads, and redownloads show very high correlations with each other (values close to 1.00), suggesting that these metrics are closely related.

- The spend metric shows moderate to high correlations with installs, impressions, taps, new downloads, and redownloads (around 0.92 to 0.94).

- The conversion rate (CR) has a low correlation with all other metrics (values around 0.23 to 0.29), indicating that it behaves independently of the volume of user interactions and spending.
